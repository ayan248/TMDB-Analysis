# Loading the dataset and brief overview

```python
import pandas as pd
movies = pd.read_csv("E:\\TMDB_movie_dataset_v11.csv")
```

```python
movies.info()
```

```python
movies.head()
```

# Cleaning the data

```python
movies = movies[['id', 'title', 'status', 'budget', 'runtime', 'vote_average', 'vote_count', 'popularity', 'revenue']]
```

```python
movies.info()
```

```python
movies.dropna(inplace=True)
```

```python
movies.info()
```

```python
movies = movies[movies['budget']>0]
movies = movies[movies['runtime']>0]
movies = movies[movies['revenue']>0]
movies = movies[movies['vote_average']>0]
movies = movies[movies['vote_count']>0]
movies = movies[movies['popularity']>0]
```

```python
movies.info()
```

# Defining features and labels

```python
features = ['budget', 'runtime', 'vote_average', 'vote_count', 'popularity']
target = 'revenue'

x = movies[features]
y = movies[target]
```

# Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=25)
```

# Standardizing Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# Linear Regressor Model

## Initializing and Training the model

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
lr_y_pred = lr_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

lr_mae = mean_absolute_error(y_test, lr_y_pred)
lr_mse = mean_squared_error(y_test, lr_y_pred)
lr_rmse = np.sqrt(lr_mse)
lr_r2 = r2_score(y_test, lr_y_pred)

print(f"Mean Absolute Error (MAE): {lr_mae}")
print(f"Mean Squared Error (MSE): {lr_mse}")
print(f"Root Mean Squared Error: {lr_rmse}")
print(f"R-squared (R2): {lr_r2}")
```

# Polynomial Regressor Model

## Initializing and Training the model

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)  
x_train_poly = poly.fit_transform(x_train_scaled)
x_test_poly = poly.transform(x_test_scaled)

poly_model = LinearRegression()
poly_model.fit(x_train_poly, y_train)
poly_y_pred = poly_model.predict(x_test_poly)
```

## Making predictions on the testing set

```python
poly_y_pred = poly_model.predict(x_test_poly)
```

## Calculating performance of the model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

poly_mae = mean_absolute_error(y_test, poly_y_pred)
poly_mse = mean_squared_error(y_test, poly_y_pred)
poly_rmse = np.sqrt(poly_mse)
poly_r2 = r2_score(y_test, poly_y_pred)

print(f"Mean Absolute Error (MAE): {poly_mae}")
print(f"Mean Squared Error (MSE): {poly_mse}")
print(f"Root Mean Squared Error: {poly_rmse}")
print(f"R-squared (R2): {poly_r2}")
```

# Decision Tree Regressor Model

## Initializing and Training the model

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
dt_y_pred = dt_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
dt_mae = mean_absolute_error(y_test, dt_y_pred)
dt_mse = mean_squared_error(y_test, dt_y_pred)
dt_rmse = np.sqrt(dt_mse)
dt_r2 = r2_score(y_test, dt_y_pred)

print(f"Mean Absolute Error (MAE): {dt_mae}")
print(f"Mean Squared Error: {dt_mse}")
print(f"Root Mean Squared Error: {dt_rmse}")
print(f"R-squared: {dt_r2}")
```

# Random Forest Regressor Model

## Initializing and training the model

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
rf_y_pred = rf_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
rf_mae = mean_absolute_error(y_test, rf_y_pred)
rf_mse = mean_squared_error(y_test, rf_y_pred)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test, rf_y_pred)

print(f"Mean Absolute Error (MAE): {rf_mae}")
print(f"Mean Squared Error: {rf_mse}")
print(f"Root Mean Squared Error: {rf_rmse}")
print(f"R-squared: {rf_r2}")
```

# Finding out the equation of the line

## For Linear Regressor Model

```python
lr_intercept = lr_model.intercept_
lr_coefficients = lr_model.coef_

print("Linear Regression Equation:")
print(f"y = {lr_intercept:.4f} + ", end="")
print(" + ".join([f"({coef:.4f})*{feat}" for coef, feat in zip(lr_coefficients, features)]))
```

## For Polynomial Regressor Model

```python
poly_intercept = poly_model.intercept_
poly_coefficients = poly_model.coef_
poly_feature_names = poly.get_feature_names_out(features)

print("\nPolynomial Regression Equation:")
print(f"y = {poly_intercept:.4f} + ", end="")
print(" + ".join([f"({coef:.4f})*{feat}" for coef, feat in zip(poly_coefficients, poly_feature_names)]))
```

# Plotting Graphs

## Actual vs Predicted Revenue

```python
import matplotlib.pyplot as plt

def plot_actual_vs_predicted(y_test, predictions, model_name):
    plt.figure(figsize=(10,6))
    plt.scatter(y_test, predictions, alpha=1, color='b')
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='r', linestyle='--') 
    plt.title(f'Actual vs Predicted Revenue ({model_name})')
    plt.xlabel('Actual Revenue')
    plt.ylabel('Predicted Revenue')
    plt.show()
```

### For Linear Regressor Model

```python
plot_actual_vs_predicted(y_test, lr_y_pred, 'Linear Regression')
```

### For Polynomial Regressor Model

```python
plot_actual_vs_predicted(y_test, poly_y_pred, 'Polynomial Regression')
```

### For Decision Tree Model

```python
plot_actual_vs_predicted(y_test, dt_y_pred, 'Decision Tree')
```

### For Random Forest Regressor Model

```python
plot_actual_vs_predicted(y_test, rf_y_pred, 'Random Forest')
```

## Feature Importance

```python
def plot_feature_importance(model, model_name, x_train_scaled):
    feature_importances = model.feature_importances_
    features = x_train.columns
    
    importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(8, 6))
    plt.barh(importance_df['Feature'], importance_df['Importance'])
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.title(f'Feature Importance - {model_name}')
    plt.gca().invert_yaxis() 
    plt.show()
```

### For Decision Tree Regressor Model

```python
plot_feature_importance(dt_model,'Decision Tree', x_train_scaled)
```

### For Random Forest Regressor Model

```python
plot_feature_importance(rf_model, 'Random Forest', x_train_scaled)
```

### For Linear Regressor Model

```python
lr_coefficients = lr_model.coef_
lr_feature_importance = pd.Series(lr_coefficients, index=features).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
lr_feature_importance.plot(kind='bar')
plt.title('Feature Importance in Linear Regression')
plt.ylabel('Coefficient Value')
plt.show()
```

### For Polynomial Regressor Model

```python
poly_coefficients = poly_model.coef_
poly_feature_names = poly.get_feature_names_out(features)
poly_feature_importance = pd.Series(poly_coefficients, index=poly_feature_names).sort_values(ascending=False)

plt.figure(figsize=(12, 6))
poly_feature_importance.plot(kind='bar')
plt.title('Feature Importance in Polynomial Regression')
plt.ylabel('Coefficient Value')
plt.show()
```
