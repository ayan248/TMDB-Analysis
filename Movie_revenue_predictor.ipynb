# Loading the dataset and brief overview

```python
import pandas as pd
movies = pd.read_csv("E:\\TMDB_movie_dataset_v11.csv")
```

```python
movies.info()
```

```python
movies.head()
```

# Cleaning the data

```python
movies = movies[['id', 'title', 'status', 'budget', 'runtime', 'vote_average', 'vote_count', 'popularity', 'revenue']]
```

```python
movies.info()
```

```python
movies.dropna(inplace=True)
```

```python
movies.info()
```

```python
movies = movies[(movies['budget'] > 0) & (movies['runtime'] > 0) & (movies['revenue'] > 0) & 
         (movies['vote_average'] > 0) & (movies['vote_count'] > 0) & (movies['popularity'] > 0)]
```

```python
movies.info()
```

# Defining features and labels

```python
features = ['budget', 'runtime', 'vote_average', 'vote_count', 'popularity']
target = 'revenue'

x = movies[features]
y = movies[target]
```

# Splitting the data into training and testing sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=25)
```

# Standardizing Scaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

# Linear Regression Model

## Initializing and Training the model

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
lr_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
lr_y_pred = lr_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
from sklearn.metrics import r2_score
import numpy as np

lr_r2 = r2_score(y_test, lr_y_pred)

print(f"R-squared (R2): {lr_r2}")
```

# Polynomial Regression Model

## Initializing and Training the model

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)  
x_train_poly = poly.fit_transform(x_train_scaled)
x_test_poly = poly.transform(x_test_scaled)

poly_model = LinearRegression()
poly_model.fit(x_train_poly, y_train)
```

## Making predictions on the testing set

```python
poly_y_pred = poly_model.predict(x_test_poly)
```

## Calculating performance of the model

```python
poly_r2 = r2_score(y_test, poly_y_pred)

print(f"R-squared (R2): {poly_r2}")
```

# Decision Tree Regression Model

## Initializing and Training the model

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor()
dt_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
dt_y_pred = dt_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
dt_r2 = r2_score(y_test, dt_y_pred)

print(f"R-squared (R2): {dt_r2}")
```

# Random Forest Regression Model

## Initializing and training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor()
rf_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
rf_y_pred = rf_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
rf_r2 = r2_score(y_test, rf_y_pred)

print(f"R-squared (R2): {rf_r2}")
```

# Gradient Boosting Regression Model

## Initializing and training the model

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=25)
gb_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
gb_y_pred = gb_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
gb_r2 = r2_score(y_test, gb_y_pred)
print(f"Gradient Boosting R-squared (R2): {gb_r2}")
```

# KNN Regression Model

## Initializing and training the model

```python
from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor(n_neighbors=11)
knn_model.fit(x_train_scaled, y_train)
```

## Making predictions on the testing set

```python
knn_y_pred = knn_model.predict(x_test_scaled)
```

## Calculating performance of the model

```python
knn_r2 = r2_score(y_test, knn_y_pred)
print(f"KNN R-squared (R2): {knn_r2}")
```

# Plotting Graphs

## Actual vs Predicted Revenue

```python
import matplotlib.pyplot as plt

def plot_actual_vs_predicted(y_test, predictions, model_name):
    plt.figure(figsize=(10,6))
    plt.scatter(y_test, predictions, alpha=1, color='b')
    plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='r', linestyle='--') 
    plt.title(f'Actual vs Predicted Revenue ({model_name})')
    plt.xlabel('Actual Revenue')
    plt.ylabel('Predicted Revenue')
    plt.show()
```

### For Linear Regression Model

```python
plot_actual_vs_predicted(y_test, lr_y_pred, 'Linear Regression')
```

### For Polynomial Regression Model

```python
plot_actual_vs_predicted(y_test, poly_y_pred, 'Polynomial Regression')
```

### For Decision Tree Regression Model

```python
plot_actual_vs_predicted(y_test, dt_y_pred, 'Decision Tree Regression')
```

### For Random Forest Regression Model

```python
plot_actual_vs_predicted(y_test, rf_y_pred, 'Random Forest Regression')
```

### For Gradient Boosting Regression Model

```python
plot_actual_vs_predicted(y_test, gb_y_pred, 'Gradient Boosting')
```

### For KNN Regression Model

```python
plot_actual_vs_predicted(y_test, knn_y_pred, 'KNN Regression')
```

## Feature Importance

```python
def plot_feature_importance(model, model_name, x_train_scaled):
    feature_importances = model.feature_importances_
    features = x_train.columns
    
    importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(8, 6))
    plt.barh(importance_df['Feature'], importance_df['Importance'])
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.title(f'Feature Importance - {model_name}')
    plt.gca().invert_yaxis() 
    plt.show()
```

### For Decision Tree Regression Model

```python
plot_feature_importance(dt_model,'Decision Tree', x_train_scaled)
```

### For Random Forest Regression Model

```python
plot_feature_importance(rf_model, 'Random Forest', x_train_scaled)
```

## R-squared vs K-Value for KNN Regression Model

```python
k_values = range(1, 20)
errors = []
for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(x_train_scaled, y_train)
    y_pred = knn.predict(x_test_scaled)
    errors.append(r2_score(y_test, y_pred))

plt.figure(figsize=(8, 6))
plt.plot(k_values, errors, marker='o', linestyle='-', color='purple')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('R-squared')
plt.title('R-squared vs K-Value')
plt.show()
```
